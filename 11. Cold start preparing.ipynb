```python
# df1 = pd.read_csv("/data/log-data-2024/NEW_MOVIE.csv")
# df2 = pd.read_csv("/data/log-data-2024/NEW_TVSHOW.csv")
df = pd.read_excel("/data/meta/2024Release_update.xlsx")

with open(file='/data/meta/241010_metadata_us_movie.pkl', mode='rb') as f:
    cf1 = pickle.load(f)
    
with open(file='/data/meta/241010_metadata_us_tvshow.pkl', mode='rb') as f:
    cf2 = pickle.load(f)
```

```python
# df = pd.concat([df1, df2], axis = 0).reset_index(drop = True)
df = df.drop_duplicates(subset='Contents Title').reset_index(drop = True)
df
```

```python
cf1 = cf1[["title_name", "genre_info", "actor_info", "director_info", "cntry_code", "description"]]
cf2 = cf2[["title_name", "genre_info", "actor_info", "director_info", "cntry_code", "description"]]
cf = pd.concat([cf1, cf2], axis = 0).reset_index(drop = True)
cf = cf.drop_duplicates(subset='title_name').reset_index(drop = True)

tn = list(df["Contents Title"])
cf["use"] = [1 if i in tn else 0 for i in cf["title_name"]]
cf = cf[cf["use"] == 1].reset_index(drop = True)

cf
```

```python
# genre -> 장르가 여러개고, 여러 언어로 기록되어 있어서 한국어만 남김
korean = re.compile("[가-힣]+")
cf["genre_info"] = [korean.findall(str(i)) for i in cf["genre_info"]]

genre_dict = {"서부" : "western", "뮤직" : "music", "로맨스" : "romance", "토크쇼예능" : "talk show",
              "어드벤처" : "adventure", "판타지" : "fantasy", "예능" : "entertainment", "리얼리티" : "reality show",
              "취미" : "hobby", "애니메이션" : "animation", "스릴러" : "thriller", "어린이" : "kids", "코미디" : "comedy",
              "에로" : "erotic", "드라마" : "drama", "액션" : "action", "시사" : "current affairs", "범죄" : "crime",
              "시사다큐" : "documentary", "뉴스" : "news", "멜로로맨스" : "melodrama", "스포츠" : "sports", "전쟁" : "war",
              "레저" : "leisure", "다큐" : "documentary", "미스터리" : "mystery", "취미레저" : "hobby", "멜로" : "melodrama",
              "호러" : "horror", "전기" : "history", "토크쇼" : "talk show", "뮤지컬" : "musical", "국제" : "international",
              "학원물" : "school"}

temp = []
for i in cf["genre_info"]:
    temp.append(list(set([genre_dict[w] for w in i])))
cf["genre_info"] = temp
```

```python
# actor -> 배우 이름 / 극중 이름으로 기재되어 있어서 배우 이름만 발라냄
cf["actor_info"] = cf["actor_info"].apply(lambda x : str(x).replace("\x0b", "").replace("\x0c", ""))

actor = re.compile("[0-9].+")
non_num = re.compile("[0-9]+")

res = []
for i in tqdm(cf["actor_info"]):
    temp = i.split("Actor")
    actors = []
    for w in temp[:3]:
        try:
            actors.append(re.sub(non_num, "", actor.findall(w)[0]))
        except:
            pass
    res.append(actors)
cf["actor_info"] = res
```

```python
# director -> 개행문자 제거
res = []
for i in cf["director_info"]:
    try:
        res.append(i.split("\x0b")[1])
    except:
        res.append(i)
cf["director_info"] = res
```

```python
cf = cf[["title_name", "genre_info", "actor_info", "director_info", "cntry_code", "description"]]
```

```python
col = list(cf.columns)[1:]

def transform(series):
    result = ""
    for a, i in enumerate(list(series.iloc[0])[1:]):
        result = result + col[a] + ": "
        try:
            if i == []:
                result = result + "None" + ", "
            elif type(i) == list:
                result = result + "["
                for w in i:
                    result = result + w + ", "
                result = result[:-2] + "]" + ", "
            else:
                result = result + i + ", "
        except:
            result = result + "None" + ", "
    return result[:-2]
```

```python
result = {}
for i in tqdm(cf["title_name"]):
    result[i] = transform(cf.loc[cf[cf["title_name"] == i].index])

result
```

```python
import pandas as pd
import torch.nn.functional as F
import pickle
from transformers import BertModel, BertTokenizer
import torch
from tqdm import tqdm
import ast

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
import ast

cold_start_n = 5 #몇개 뽑고 싶니?

with open(file='/data/log-data-2024/0.version checkpoints/20241022/match_dict_final2.pickle', mode='rb') as f:
    dic = pickle.load(f) # 콘텐츠 아이디

with open(file='/data/log-data-2024/0.version checkpoints/20241022/content_meta_dict_final.pickle', mode='rb') as f:
    meta = pickle.load(f) # meta
    
prediction = pd.read_csv("/data/log-data-2024/8man_sample_20241022_treatment2.csv")
prediction["treatment2 prediction"] = prediction["treatment2 prediction"].apply(lambda x : ast.literal_eval(x))
prediction = list(prediction["treatment2 prediction"])

# with open(file='/data/log-data-2024/yh/LLM_EB/src/LLM_for_Embeddings_Preidction.pickle', mode='rb') as f:
#     prediction = pickle.load(f) # prediction
```

```python
df = pd.DataFrame({'CONTENTS_ID' : list(cf["contents_id"]), 'CONTENTS_SET_ID' : list(cf["contents_set_id"]), 'description' : list(result.values()), 'TITLE_NAME' : list(result.keys())})
df
```

```python
#new contents 텐서 dict 만들기 ================description 컬럼에 맞게 변경

new_contnets_vector ={}
for title, desc in zip(tqdm(df['TITLE_NAME']), df['description']):
    inputs = tokenizer(desc, padding="max_length", truncation=True, return_tensors="pt", max_length=512 , add_special_tokens=False)
    with torch.no_grad():
        outputs = model(**inputs) 
        embeddings = outputs[0]
    new_contnets_vector[title] = embeddings
        
#추가된 새로운 데이터
new_value = list(new_contnets_vector.values())
new_key = list(new_contnets_vector.keys())
```

```python
#콘텐츠 아이디 키값 아이디 스위치
content_dic = {int(value): key for key, value in dic.items()}
content_dic
```

```python
final_cold_result = []

for pred in tqdm(prediction):
    
    pred = pred[:5]
    
    vector_stack = []
    for con_id in pred:
        #meta key값 형식 만들기 (변수 key)
        title = content_dic[con_id]
        key = str(con_id) + '. ' +  str(title)
        try:
            value = meta[key] 
        except:
            print(key)
            pass

        inputs = tokenizer(value, padding="max_length", truncation=True, return_tensors="pt", max_length=512 , add_special_tokens=False)
        with torch.no_grad():
            outputs = model(**inputs) 
            embeddings = outputs[0]

        vector_stack.append(embeddings)

    #평균값 구하기
    vector_stack = torch.stack(vector_stack) 
    mean_tensor = torch.mean(vector_stack, dim=0)


    #유사도구하기
    similarity_dict = {}
    for i, n_key, n_value in zip(list(range(len(new_key))), new_key, new_value):
        cosine_similarity = F.cosine_similarity(mean_tensor, n_value, dim=-1)
        cosine_similarity = cosine_similarity.mean()
        similarity_dict[n_key] = round(cosine_similarity.item(),6)

    #내림차순 정렬
    similarity_dict = dict(sorted(similarity_dict.items(), key=lambda item: item[1], reverse=True)) #유사도 딕셔너리 정렬
    cold_contents_title = list(similarity_dict.keys()) #key값 정렬
    cold_start_result = cold_contents_title[0:cold_start_n]

    final_cold_result.append(cold_start_result)
```

```python
final = pd.read_csv("/data/log-data-2024/8man_sample_20241022_treatment2.csv")
final["treatment2 cold contents"] = final_cold_result
final.to_csv("/data/log-data-2024/8man_sample_20241022_treatment2.csv")
```
