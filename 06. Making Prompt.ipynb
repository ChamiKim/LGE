- LLM as RS를 위해서 프롬프트를 구축합니다.

```python
import pickle
import pandas as pd
from tqdm import tqdm

with open(file='/data/log-data-2024/2.sequence_generate_ksc/data/match_dict_final.pickle', mode='rb') as f:
    dic = pickle.load(f)
    
with open(file='/data/log-data-2024/2.sequence_generate_ksc/data/content_meta_dict_final.pickle', mode='rb') as f:
    cnt_meta = pickle.load(f)

df = pd.read_csv("/data/log-data-2024/Final_20241127/input_search_augmented_final_20241127.txt", sep = "\t", header = None)
meta = pd.read_csv("/data/log-data-2024/Final_20241127/Meta_20241127_optimized.csv")
    
with open(file = '/data/log-data-2024/Final_20241127/match_dict_final2.pickle', mode='rb') as f:
    dic = pickle.load(f)
```

```python
cf = []
for i in tqdm(df[0]):
    temp = i.split()[1:]
    cf2 = []
    for w in temp:
        cf2.append(w + ". " + str(dic[w]))
    cf.append(cf2)
cf
```

```python
def make_prompt(lit):
    lit = lit[:-1]
    return f"You are a recommender system. Based on the contents that the user has previously chosen, recommend a proper TV content that the user may prefer. Output the title of the recommended TV content which is listed in the dictionary. Previously chosen contents: {lit}."

result = [make_prompt(i) for i in cf]

output = [i[-1] for i in cf]

prompt = []
for a, i in enumerate(result):
    prompt.append({"prompt" : i, "completion" : output[a]})
prompt
```
